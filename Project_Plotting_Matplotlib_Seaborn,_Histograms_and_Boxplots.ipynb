```python
import pandas as pd
data = pd.read_csv('fuel_econ.csv')


print(data.head(6))
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['comb'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Fuel Efficiency')
plt.xlabel('Fuel Efficiency (mpg)')
plt.ylabel('Frequency')
plt.savefig('Distribution of Fuel Efficiency.png')
plt.show()
```

# Interpretation for the Histogram:
# The histogram visualizes the distribution of fuel efficiency (measured in mpg) across the dataset.
# This shows that most vehicles have fuel efficiency in the range of 20 to 30 mpg, indicating that the majority fall
# into a moderate fuel efficiency category. There are very few vehicles with extremely high mpg (greater than 40).
# This suggests that most vehicles in the dataset are conventional cars, rather than hybrid orelectricones.

```python
plt.figure(figsize=(12, 8))
numerical_data = data.select_dtypes(include=np.number)
correlation_matrix = numerical_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.savefig('Feature Correlation Heatmap.png')
plt.show()
```

# Interpretation for the Heatmap:
# The heatmap shows the correlation between different features in the dataset:
# - Strong positive correlation is observed between co2 and displ,
#   indicating that vehicles with larger engine displacement emit more CO2.
# - Similarly, city and highway are strongly correlated, suggesting
#   vehicles with good city mileage also perform well on highways.
# on the right side of the heat map from city to ghgscore excluding co2 all have strong positive correlation
# - Features such as id and cylinders or displ have weak or no correlation,
#   implying that these pairs are largely independent of each other.
# - Negative correlations, such as between co2 and fuel efficiency (mpg), are
#   intuitive as vehicles emitting more CO2 tend to have lower fuel efficiency.
# Overall, these insights can guide feature selection for AI/ML models, focusing
# on strongly correlated variables that influence fuel efficiency orCO2emissions.

```python
# Save the histogram
plt.savefig('Distribution of Fuel Efficiency.png')

# Save the heatmap
plt.savefig('Feature Correlation Heatmap.png')
```
